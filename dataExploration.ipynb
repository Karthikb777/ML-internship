## Handling NaN values in the dataset

### We handle NaN values by checking their dtype:
### if the dtype is object, we fill it with NA i.e Not Available
### else we fill it with 0

```python
cols = raw_data.columns.to_list()
cpy = raw_data.copy()
for col in cols:
    # cpy = cpy.dropna()
    cpy = cpy.fillna(method='pad')

# raw_data.isnull() - raw_data still has NaN values because it is untouched
cpy.isnull()


# # we have made a copy of the raw_data to process it and use it for our ML algorithm, after filling NaN values, cpy doesn't have any NaN values as seen below 
# cpy

cpy.isnull().sum()
```

### We have many columns that are not required such as Customer ID, Name etc. We are going to drop all those unnecessary fields from the dataset. The gender column is also dropped because we are building a model that will not be gender biased while predicting the loan amount to be sanctioned.

```python
cpy2 = cpy
cpy = cpy.drop(columns=['Customer ID', 'Name', 'Gender', 'Profession', 'Property ID', 'Has Active Credit Card'])
cpy2 = cpy2.drop(columns=['Customer ID', 'Name', 'Gender', 'Profession', 'Property ID', 'Type of Employment', 'Expense Type 1', 'Expense Type 2', 'Property Type', 'Property Location'])
cpy2
```
