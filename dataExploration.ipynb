## Handling NaN values in the dataset

### We handle NaN values by checking their dtype:
### if the dtype is object, we fill it with NA i.e Not Available
### else we fill it with 0

```python
cols = raw_data.columns.to_list()
cpy = raw_data.copy()
for col in cols:
    if cpy[col].dtype == 'O':
        # do fillna('NA') here
        cpy[col].fillna('NA', inplace=True)
    else:
        # do fillna(0) here
        cpy[col].fillna(0, inplace=True)

# raw_data.isnull() - raw_data still has NaN values because it is untouched
cpy.isnull() # we have made a copy of the raw_data to process it and use it for our ML algorithm, after filling NaN values, cpy doesn't have any NaN values as seen below 
```
