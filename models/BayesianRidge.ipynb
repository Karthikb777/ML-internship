```python
import pandas as pd
import numpy as np
```

```python
# importing the training dataset
data = pd.read_csv('../encoded data/enc3.csv')
data
```

```python
from sklearn.linear_model import BayesianRidge
BRmodel = BayesianRidge()
```

```python
X = data.drop(['Loan Sanction Amount (USD)', 'Unnamed: 0'], axis=1)
X
```

```python
y = data['Loan Sanction Amount (USD)']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
# training the model
BRmodel.fit(X_train, y_train)
```

```python
# testing the model
predictions = BRmodel.predict(X_test)
```

```python
score = BRmodel.score(X_test, y_test)
score


from sklearn.metrics import r2_score
r2_score(y_test, predictions)
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, predictions)
plt.xlabel('test data')
plt.ylabel('predicted data')

plt.show()
```
